# Schedule Planner 2: Understanding the USF Staff Search Application

This document aims to decode the HTTP POST created by the [USF Staff search](https://usfweb.usf.edu/DSS/StaffScheduleSearch), a publicly accessible application that allows a public user to search for classes based off of the following criteria: 

[Semester ```P_SEMESTER```](#semester)

[Session ``P_SESSION``](#session)

[Campus ``P_CAMPUS``](#session)

[College ``P_COL``](#college)

[Department ``P_DEPT``](#department)

[Course Status ``p_status``](#status)

[Actual Course Status ``p_ssts_code``](#actual_status)

[Course Level ``P_CRSE_LEVL``](#level)

[Course Reference Number ``P_REF``](#crn)

[Course Subject ``P_SUBJ``](#subject)

[Course Number ``P_NUM``](#number)

[Course Title ``P_TITLE``](#title)

[Course Credit Hours ``P_CR``](#credit_hours)

[Instructional Method ``p_insm_x_inad``, ``p_insm_x_incl``, ``p_insm_x_inhb``, ``p_insm_x_inpd``, ``p_insm_x_innl``, ``p_insm_x_inot``](#instructional_method)

[Class Meets Only on ... ``p_day_x``, ``p_day_xm``, ``p_day_xt``, ``p_day_xw``, ``p_day_xr``, ``p_day_xt``, ``p_day_xf``, ``p_day_xs``, ``p_day_xu``](#class_meets_only_on)

[Class Must Meet At Least On ... ``p_day``, ``p_daym``, ``p_dayt``, ``p_dayw``, ``p_dayr``, ``p_dayt``, ``p_dayf``, ``p_days``, ``p_dayu``](#class_must_meet_at_least_on)

[Class Must Start at ... ``P_TIME1``](#class_must_start_at)

[Instructor Name ``P_INSTRUCTOR``](#instructor_name)

[Degree Program Attributes ``P_UGR``](#degree_program_attributes)


Steps taken to acquire this data:
1. Navigate to the [USF Staff search application](http://usfweb.usf.edu/DSS/StaffScheduleSearch)
2. Ensure you are visiting the website via http (NOT HTTPS) and open [wireshark](https://www.wireshark.org/download.html) to capture your network traffic
3. Find the HTTP POST made by the application
4. Match parameters found in the HTTP POST to IDs found on the form


The following is the full default curl request for requesting all classes from Fall 2024

````python
curl -X POST "http://usfweb.usf.edu/DSS/StaffScheduleSearch/StaffSearch/Results" \
     -H "Connection: keep-alive" \
     -H "Content-Length: 294" \
     -H "Cache-Control: max-age=0" \
     -H "Upgrade-Insecure-Requests: 1" \
     -H "User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36 OPR/107.0.0.0" \
     -H "Origin: http://usfweb.usf.edu" \
     -H "Content-Type: application/x-www-form-urlencoded" \
     -H "Accept: text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7" \
     -H "Referer: http://usfweb.usf.edu/DSS/StaffScheduleSearch/" \
     -H "Accept-Encoding: gzip, deflate" \
     -H "Accept-Language: en-US,en;q=0.9" \
     --data-urlencode "P_SEMESTER=202408" \
     --data-urlencode "P_SESSION=" \
     --data-urlencode "P_CAMPUS=" \
     --data-urlencode "P_COL=" \
     --data-urlencode "P_DEPT=" \
     --data-urlencode "p_status=" \
     --data-urlencode "p_ssts_code=" \
     --data-urlencode "P_CRSE_LEVL=" \
     --data-urlencode "P_REF=" \
     --data-urlencode "P_SUBJ=" \
     --data-urlencode "P_NUM=" \
     --data-urlencode "P_TITLE=" \
     --data-urlencode "P_CR=" \
     --data-urlencode "p_insm_x_inad=YAD" \
     --data-urlencode "p_insm_x_incl=YCL" \
     --data-urlencode "p_insm_x_inhb=YHB" \
     --data-urlencode "p_insm_x_inpd=YPD" \
     --data-urlencode "p_insm_x_innl=YNULL" \
     --data-urlencode "p_insm_x_inot=YOT" \
     --data-urlencode "p_day_x=no_val" \
     --data-urlencode "p_day=no_val" \
     --data-urlencode "P_TIME1=" \
     --data-urlencode "P_INSTRUCTOR=" \
     --data-urlencode "P_UGR="
````

The following is the simplest HTTP POST that will return all sections of ENC 1101 for the Fall 2024 semester.

The headers used are 

[Semester ```P_SEMESTER```](#semester)
